```python
df_org=pd.read_csv("./accidentinfo.csv", sep=',')
```

```python
df_org.drop('사고내용',axis=1,inplace=True)
df_org
```

```python
df_org['시군구']=df_org['시군구'].str.strip(to_strip='서울특별시')
df_org
```

```python
c1=df_org['시군구'].str.split().tolist()
result1=[]
result2=[]
for i in range(len(c1)):
    goo=c1[i][0]
    dong=c1[i][1]
    result1.append(goo)
    result2.append(dong)
len(result1), len(result2)
```

```python
df_org['자치구']=result1
df_org['행정구']=result2
```

```python
df_org
```

```python
df_org.drop('시군구',axis=1,inplace=True)
```

```python
df=df_org[['자치구','행정구','사망자수','중상자수','경상자수','도로형태','가해운전자 차종','피해운전자 차종']]
```

```python
df=df.loc[df['가해운전자 차종' or '피해운전자 차종']=='이륜']
```

```python
df
```

```python
road_split=df.도로형태.str.split(' - ')

road_split.index=range(0,3545)

print(road_split)

print(type(road_split))
print(type(road_split[0]))

#단일로, 교차로만 데이터로 넣기
road=road_split.str.get(0)

#필요없는 열 삭제
df.drop(['가해운전자 차종','피해운전자 차종'], axis=1,inplace=True)

df

df.drop(['도로형태'],axis=1,inplace=True)

df.index=range(3545)

df['도로형태']=road
```

```python
df
```

```python
sglroad= df['도로형태']=='단일로'
crssroad= df['도로형태']=='교차로'
df['단일로']=sglroad
df['교차로']=crssroad
df['단일로']=df['단일로'].map(lambda x: 1 if x else 0)
df['교차로']=df['교차로'].map(lambda x: 1 if x else 0)
```

```python
df
```

```python
#컬럼명 변환
df=df.rename(columns={'행정구':'행정동'})
```

```python
#행정동과 법정동을 매칭시켜주는 파일을 불러옴
df_match= pd.read_csv('./행정동 법정동 코드.csv', encoding='utf-8')
```

```python
df_match
```

```python
#서울시, 종로구 등 시군구 단위까지만 있는 것은 값이 없기 때문에 결측치로 처리됨 : 해당 결측치 제거
df_match=df_match.dropna()
```

```python
df_match
```

```python
#행정동/법정동 코드를 10자리->8자리로 변경
df_match['행정동코드']=(df_match['행정동코드']/100).astype(np.int64)
df_match['법정동코드']=(df_match['법정동코드']/100).astype(np.int64)
```

```python
#하나의 행정동이 몇개의 법정동으로 이루어져 있는지 확인
bup_count=df_match[['행정동코드','법정동코드']].groupby('행정동코드').count().rename(columns={'법정동코드':'법정동갯수'}).reset_index()
bup_count
```

```python
#df_match에 추가
df_match=pd.merge(df_match,bup_count,how='left').loc[df_match['시도명']=='서울특별시',:]
df_match
```

```python
#서울시 행정동별 인구데이터 불러오기
df_stat=pd.read_csv('C:/Users/김다은/Desktop/개인분석과제/서울시 행정동별 생활인구(내국인).csv', encoding='utf-8', index_col=None)
```

```python
df_stat
```

```python
#행정동-법정동 매칭 정보와 통계치를 붙임
df2=pd.merge(df_match,df_stat,how='left')
df2.dropna()
```
