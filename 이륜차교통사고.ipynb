```python
c1=df_org['시군구'].str.split().tolist()
result1=[]
result2=[]
for i in range(len(c1)):
    goo=c1[i][0]
    dong=c1[i][1]
    result1.append(goo)
    result2.append(dong)
len(result1), len(result2)
```
