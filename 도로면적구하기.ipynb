```python
import pandas as pd
```

```python
df_road=pd.read_csv('도로현황(도로율).csv', encoding='utf-8')
df_area=pd.read_csv('서울시 행정동별 면적.csv',encoding='utf-8')
```

```python
df_road.drop('자치구별(1)', axis=1, inplace=True)
```

```python
df_road.columns=['자치구','도로율(%)']
```

```python
df_road.drop(0,axis=0, inplace=True)
```

```python
df_road.drop(1,axis=0,inplace=True)
```

```python
df_road.index=range(25)
```

```python
type(np.array(df_road['도로율(%)'])[0])   #str 을 float 로 바꿔야함
```

```python
df_road['도로율(%)']=df_road['도로율(%)'].astype(float) #str->float 변환
```

```python
df_road['도로율(%)'].dtype
```

```python
df_road['도로율(%)']=df_road['도로율(%)']*0.01 
```

```python
df_road.rename(columns={"도로율(%)": "도로율"},inplace=True) # 컬럼명 변경
```

```python
df_area
```

```python
df_area.drop('동별(1)',axis=1, inplace=True)
df_area.drop([0,1,2], axis=0,inplace=True)
```

```python
df_area.columns=['자치구','행정동','면적(k㎡)']
```

```python
df_area=df_area.loc[df_area['행정동']!='소계']
```

```python
df_area.index=range(426)
```

```python
df_area
```

```python
# merge 하기
df_area.merge(df_road, how='left', on='자치구').sort_values(by='자치구',ascending=True)
```

```python
df_merged=df_area.merge(df_road, how='left', on='자치구')
```

```python
#도로면적 구하기 (행정동별면적 * 구별 도로율 )
df_merged['면적(k㎡)']*df_merged['도로율']
```

```python
#str->float 변환
print(df_merged['면적(k㎡)'].dtype)
print(df_merged['도로율'].dtype)
```

```python
df_merged['면적(k㎡)']=df_merged['면적(k㎡)'].astype(float)
```

```python
#도로면적 계산후 새 컬럼에 추가
df_merged['도로면적']=df_merged['면적(k㎡)']*df_merged['도로율']
```

```python
df_com=df_merged[['행정동','도로면적']]
```

```python
df_com
```

```python
df_com.to_csv('행정동_도로면적(km2).')
```
